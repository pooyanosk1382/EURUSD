```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('1H.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df = df.loc[~(df['last open10'].isnull())]
df
```

```python
x = df.drop(columns=['<HIGH>', '<LOW>', '<CLOSE>'])
y = df['<HIGH>']
x
```

```python
regressorHigh = XGBRegressor(tree_method='hist', n_estimators=2000)
regressorHigh.fit(x.values, y.values)
predHigh = np.around(regressorHigh.predict(x.values), 4)
predHigh
```

```python
x = df.drop(columns=['<LOW>', '<CLOSE>'])
y = df['<LOW>']
x
```

```python
regressorLow = XGBRegressor(tree_method='hist', n_estimators=2000)
regressorLow.fit(x.values, y.values)
predLow = np.around(regressorLow.predict(x.values), 4)
```

```python
predLow
```

```python
x = df.drop(columns=['<CLOSE>'])
y = df['<CLOSE>']
x
```

```python
regressorClose = XGBRegressor(tree_method='hist', n_estimators=2000)
regressorClose.fit(x.values, y.values)
predClose = np.around(regressorClose.predict(x.values), 4)
predClose
```

```python
testSet = pd.read_csv('test.csv')
testSet
```

```python
testSet.drop(columns='Unnamed: 0', inplace=True)
testSet
```

```python
testDF = pd.DataFrame()
```

```python
testDF['Open'] = testSet['<OPEN>']
testDF['Real Close'] = testSet['<CLOSE>']
testDF
```

```python
xx = testSet.iloc[:, :34]
xx['predHigh'] = np.around(regressorHigh.predict(xx.values), 5)
xx
```

```python
xx['predLow'] = np.around(regressorLow.predict(xx.values), 5)
xx
```

```python
xx['predClose'] = np.around(regressorClose.predict(xx.values), 5)
xx
```

```python
testDF['predClose'] = xx['predClose']
testDF['Real Close - Open'] = testDF['Real Close'] - testDF['Open']
testDF['predClose - Open'] = testDF['predClose'] - testDF['Open']
testDF['product'] = testDF['Real Close - Open'] * testDF['predClose - Open']
testDF
```
