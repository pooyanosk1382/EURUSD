```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from xgboost import XGBRegressor
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv("4H.csv")
dataset = dataset.iloc[11:, 1:]*10**5 - 131904
openVal = dataset.iloc[:, 33]
closeVal = dataset.iloc[:, 36]
openVal = openVal
closeVal = closeVal
openVal = openVal.values
closeVal = closeVal.values
test = []
for i in range(len(openVal)):
    if closeVal[i] - openVal[i] < 0:
        test.append(0)
    else:
        test.append(1)
print(openVal)
print(closeVal)
print(test)
```

```python
x = dataset.iloc[:, 1:34].values
y = dataset.iloc[:,   34].values
y = y.reshape(len(y),1)
regressorHigh = XGBRegressor()
regressorHigh.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 131904
yHigh = regressorHigh.predict(testSet.iloc[:, 1:34].values)
yHigh += 131904
yHigh /= 10**5
tempHigh = np.concatenate((yHigh.reshape(len(yHigh), 1), (testSet.iloc[:, 34:35] + 131904)/10**5), 1)
tempHigh
```

```python
from sklearn.metrics import r2_score
r2_score((testSet.iloc[:, 34:35] + 131904)/10**5, yHigh)
```

```python
x = dataset.iloc[:, 1:35].values
y = dataset.iloc[:,   35].values
y = y.reshape(len(y),1)
regressorLow = XGBRegressor()
regressorLow.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 131904
temp = np.concatenate((testSet.iloc[:, 1:34].values, (yHigh*10**5 - 131904).reshape(len(yHigh), 1)), 1)
temp
```

```python
yLow = regressorLow.predict(temp)
yLow += 131904
yLow /= 10**5
tempLow = np.concatenate((yLow.reshape(len(yLow), 1), (testSet.iloc[:, 35:36] + 131904)/10**5), 1)
tempLow
```

```python
from sklearn.metrics import r2_score
r2_score((testSet.iloc[:, 35:36] + 131904)/10**5, yLow)
```

```python
x = dataset.iloc[:, 1:36].values
y = dataset.iloc[:,   36].values
y = y.reshape(len(y),1)
regressorClose = XGBRegressor()
regressorClose.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 131904
temp = np.concatenate((testSet.iloc[:, 1:34].values, (yHigh*10**5 - 131904).reshape(len(yHigh), 1)), 1)
temp = np.concatenate((temp, (yLow*10**5 - 131904).reshape(len(yLow), 1)), 1)
temp
```

```python
yClose = regressorClose.predict(temp)
yClose += 131904
yClose /= 10**5
tempClose = np.concatenate((yClose.reshape(len(yClose), 1), (testSet.iloc[:, 36:37] + 131904)/10**5), 1)
tempClose
```

```python
from sklearn.metrics import r2_score
r2_score((testSet.iloc[:, 36:37] + 131904)/10**5, yClose)
```

```python
y = test
classifierAscending = XGBClassifier(n_estimators=2000)
classifierAscending.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, 1:]*10**5 - 131904
openVal = testSet.iloc[:, 33].values
closeVal = testSet.iloc[:, 36].values
ascending = []
for i in range(len(openVal)):
    if closeVal[i] - openVal[i] < 0:
        ascending.append(0)
    else:
        ascending.append(1)
ascending
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, 1:]*10**5 - 131904
temp = np.concatenate((testSet.iloc[:, 1:34].values, (yHigh*10**5 - 131904).reshape(len(yHigh), 1)), 1)
temp = np.concatenate((temp, (yLow*10**5 - 131904).reshape(len(yLow), 1)), 1)
yAscending = classifierAscending.predict(temp)
yAscending
```

```python
cm = confusion_matrix(ascending, yAscending)
print(cm)
accuracy_score(ascending, yAscending)
```
