```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from xgboost import XGBRegressor
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv("1H.csv")
dataset = dataset.iloc[22:, :]*10**5 - 127868.5
openVal = dataset.iloc[:, 67]
closeVal = dataset.iloc[:, 70]
openVal = openVal
closeVal = closeVal
openVal = openVal.values
closeVal = closeVal.values
test = []
for i in range(len(openVal)):
    if closeVal[i] - openVal[i] < 0:
        test.append(0)
    else:
        test.append(1)
print(openVal)
print(closeVal)
print(test)
```

```python
x = dataset.iloc[:, 1:68].values
y = dataset.iloc[:,   68].values
y = y.reshape(len(y),1)
regressorHigh = XGBRegressor()
regressorHigh.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 127868.5
yHigh = regressorHigh.predict(testSet.iloc[:, 1:68].values)
yHigh += 127868.5
yHigh /= 10**5
tempHigh = np.concatenate((yHigh.reshape(len(yHigh), 1), (testSet.iloc[:, 68:69] + 127868.5)/10**5), 1)
tempHigh
```

```python
from sklearn.metrics import r2_score
r2_score((testSet.iloc[:, 68:69] + 127873)/10**5, yHigh)
```

```python
x = dataset.iloc[:, 1:69].values
y = dataset.iloc[:,   69].values
y = y.reshape(len(y),1)
regressorLow = XGBRegressor(n_estimators=2000)
regressorLow.fit(x, y)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 127868.5
temp = np.concatenate((testSet.iloc[:, 1:68].values, (yHigh*10**5 - 127868.5).reshape(len(yHigh), 1)), 1)
temp
```

```python
yLow = regressorLow.predict(temp)
yLow += 127868.5
yLow /= 10**5
tempLow = np.concatenate((yLow.reshape(len(yLow), 1), (testSet.iloc[:, 69:70] + 127868.5)/10**5), 1)
tempLow
```

```python
from sklearn.metrics import r2_score
r2_score((testSet.iloc[:, 69:70] + 127873)/10**5, yLow)
```

```python
testSet = pd.read_csv('test.csv')
testSet = testSet.iloc[:, :]*10**5 - 127868.5
temp = np.concatenate((testSet.iloc[:, 1:68].values, (yHigh*10**5 - 127868.5).reshape(len(yHigh), 1)), 1)
temp = np.concatenate((temp, (yLow*10**5 - 127868.5).reshape(len(yLow), 1)), 1)
temp
```
